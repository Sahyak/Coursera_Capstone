### To Import The Covid 19 Dataset of INDIA

```python
df=pd.read_csv('covid_19_india.csv')
# To display Dataset
df
```

```python
# State Wise number of 'Cured', 'Dead' and 'Confirmed' patients
Data=df.groupby(by='State/UnionTerritory').sum()[['Cured','Deaths','Confirmed']]
Data
```

```python
# The Given Bar Graph shows you State Wise Patients in India. 
#The Yellow Bar is for Confirmed patients over time in each state. 
#Likewise Green and Read Bars are for Cured and Dead patients respectively.

df.groupby(by='State/UnionTerritory').sum().sort_values(by='Confirmed',ascending=False)[['Cured','Deaths','Confirmed']].plot(kind='bar',figsize=(20,8),color=['green','red','orange'])
plt.xlabel("States",fontsize=18)
plt.show()

# I have plotted the Bar graph in descending order as per confirmed patients. As you can see patients are more in Maharashtra State.
```

```python
# Now I will only consider the data related to Maharashtra State
mahadata=df[df['State/UnionTerritory']=="Maharashtra"].drop(['ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
mahadata
```

```python
# Here in this line graph we can see how the confirmed cases were increased in maharashtra over the time
# Given graph is expontial in nature
mahadata.plot(kind='line',x='Date',y='Confirmed',figsize=(20,8))
plt.xlabel('DATE')
plt.ylabel('CONFIRMED')
plt.show()
```

```python
# Here in this line graph we can see how the confirmed cases were increased in Gujarat over the time
# Given graph is Linear in nature
gujdata=df[df['State/UnionTerritory']=="Gujarat"].drop(['ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
gujdata.plot(kind='line',x='Date',y='Confirmed',figsize=(20,8))
plt.xlabel('DATE')
plt.ylabel('CONFIRMED')
plt.show()
```
